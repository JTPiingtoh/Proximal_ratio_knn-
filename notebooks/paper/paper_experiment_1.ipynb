```python
# for fetching datasets
from ucimlrepo import fetch_ucirepo 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tqdm import tqdm

from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import cross_validate

# the custom model
from PRKNNClassifier.test_class import PRKNeighborsClassifier

from figure_helpers import generate_score_plot
```

```python
# for sampling
RANDOM_STATE = 33
```

```python
uciml_repos_name_id_dict = {
    "Banknote": 267,
    "Glass": 42,
    "Ionosphere": 52,
    "Wine": 109,
    "Parkisons": 174,
    "Sonar": 151,
    "Haberman": 43,
    "Page blocks": 78,
    "Letter recognition": 59,
    "Ecoli": 39,
    "Optical digits": 80,
    "Pen digits": 81,
    "Transfusion": 176,
    "Musk2": 75
}
```

```python
macro_fig, macro_ax = plt.subplots(nrows=7, ncols=2, sharex=True,  figsize=(10, 30))
macro_fig.suptitle("F1_macro")
macro_axs = macro_ax.ravel()

micro_fig, micro_ax = plt.subplots(nrows=7, ncols=2, sharex=True,  figsize=(10, 30))
micro_fig.suptitle("F1_micro")
micro_axs = micro_ax.ravel()

weighted_fig, weighted_ax = plt.subplots(nrows=7, ncols=2, sharex=True,  figsize=(10, 30))
weighted_fig.suptitle("F1_weighted")
weighted_axs = weighted_ax.ravel()

k_values = [5, 15, 30, 45]

scoring = ('f1_macro', 'f1_micro', 'f1_weighted')


for i, dataset_name_id in tqdm(enumerate(uciml_repos_name_id_dict.items())):

    dataset_name, id = dataset_name_id
    print(f"Working on {dataset_name}")

    # fetching the data
    print("fetching data")
    dataset = fetch_ucirepo(id=id)
    X = dataset.data.features
    y = dataset.data.targets

    print(f"Size: {X.shape}")

    # # TODO: remove after testing loops
    # X = X.sample(50)
    # y = y.sample(50)

    # fit with k_values and plot reslts

    print("transforming target")
    if y.dtypes.iloc[0] == 'object':
        y = LabelEncoder().fit_transform(y)
    else:
        y=np.ravel(y)

    mnmx_scaler = MinMaxScaler()
    X = mnmx_scaler.fit_transform(X)

    standard_scores = []
    weighted_scores = []
    enhanced_scores = []

    
    for k in k_values:
        
        print(f"n_k={k}")

        print(f"fitting standard model")
        pr_knn_standard = PRKNeighborsClassifier(
            pr_version='standard',
            base_knn_params={"n_neighbors": k}
            ).fit(X, y)
        
        print(f"fitting weighted model")
        pr_knn_weighted = PRKNeighborsClassifier(
            pr_version='weighted',
            base_knn_params={"n_neighbors": k}
            ).fit(X,y)
        
        print(f"fitting enhanced model")
        pr_knn_enhanced = PRKNeighborsClassifier(
            pr_version='enhanced',
            base_knn_params={"n_neighbors": k}
            ).fit(X,y)

        print("CVing standard model")
        standard_scores.append(
                cross_validate(
                estimator=pr_knn_standard,
                X=X,
                y=y,
                scoring=scoring,
                cv=10
                )
            )
        
        print("CVing weighted model")
        weighted_scores.append(
                cross_validate(
                estimator=pr_knn_weighted,
                X=X,
                y=y,
                scoring=scoring,
                cv=10
                )
        )

        print("CVing enhanced model")
        enhanced_scores.append(
                cross_validate(
                estimator=pr_knn_enhanced,
                X=X,
                y=y,
                scoring=scoring,
                cv=10
                )
        )

    f1_micro_scores = {}
    f1_macro_scores = {}
    f1_weighted_scores = {}


    f1_micro_scores["standard"] = np.array([d['test_f1_micro'] for d in standard_scores])
    f1_micro_scores["enhanced"] = np.array([d['test_f1_micro'] for d in enhanced_scores])
    f1_micro_scores["weighted"] = np.array([d['test_f1_micro'] for d in weighted_scores])

    f1_macro_scores["standard"] = np.array([d['test_f1_macro'] for d in standard_scores])
    f1_macro_scores["enhanced"] = np.array([d['test_f1_macro'] for d in enhanced_scores])
    f1_macro_scores["weighted"] = np.array([d['test_f1_macro'] for d in weighted_scores])

    f1_weighted_scores["standard"] = np.array([d['test_f1_weighted'] for d in standard_scores])
    f1_weighted_scores["enhanced"] = np.array([d['test_f1_weighted'] for d in enhanced_scores])
    f1_weighted_scores["weighted"] = np.array([d['test_f1_weighted'] for d in weighted_scores])

    

    generate_score_plot(
        scores=f1_macro_scores,
        metric= "F1",
        k_values=k_values,
        dataset_name=dataset_name,
        ax=macro_axs[i]
    )

    generate_score_plot(
        scores=f1_micro_scores,
        metric= "F1",
        k_values=k_values,
        dataset_name=dataset_name,
        ax=micro_axs[i]
    )

    generate_score_plot(
        scores=f1_weighted_scores,
        metric= "F1",
        k_values=k_values,
        dataset_name=dataset_name,
        ax=weighted_axs[i]
    )
```
